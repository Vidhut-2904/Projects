# Step8: If for any column(s), the variance is equal to zero, then you need to remove those variable(s).
# Steps 9: Make sure the data is now changed into numericals

#### Step 8.1 ( apply label encoder) 

```python
### OF TRAIN DATA
def enco(dataframe):
    enco = LabelEncoder()
    for column in dataframe.columns:
        if dataframe[column].dtype == 'object':  
            dataframe[column] = enco.fit_transform(dataframe[column])
    return dataframe

X_encoded = pd.DataFrame
X_encoded = enco(X)
```

```python
X_encoded ## encoded X of train 
```

```python
### variance =0 removal
def zvarc(dataf):
    varcs = dataf.var()
    zolumns = varcs[varcs == 0].index
    cleaned_data = dataf.drop(columns=zolumns)
    return cleaned_data

cleaned_X = zvarc(X_encoded)
cleaned_X.astype(int)  ### changing data type to int   
```

```python
cleaned_X
```

```python
### OF TEST DATA
datatest_encoded = enco(datatest) ### encoded the data
datatest_encoded
```

```python
cleaned_datatest = zvarc(datatest_encoded)
cleaned_datatest ### variance =0 removal
cleaned_datatest.astype(int)  ### changing data type to int   
```

```python
cleaned_datatest.dtypes
```

# Step10: Perform dimensionality reduction

#### step 10.1 (Linear dimensionality reduction using svd)


```python
## of train dataset
svd = TruncatedSVD(n_components = 1)
reduced_X = svd.fit_transform(cleaned_X)
reduced_X = pd.DataFrame(reduced_X, columns=['Dimension1'])
```

```python
reduced_X
```

```python
### of test dataset
svd2 = TruncatedSVD(n_components = 1)
reduced_datatest = svd2.fit_transform(cleaned_datatest)
reduced_datatest = pd.DataFrame(reduced_datatest, columns=['Dimension1'])
reduced_datatest
```

# Step11: Training using xgboost

# Step12: Predict your test_df values using xgboost

```python
X_train = np.array(reduced_X)  ### all the X columns from training data
Y_train = Y.sort_values(by='y', ascending=True)##ascending order for xgboost         ### the Y column from the training data
Y_train = np.array(Y_train)
Y_test = np.array(reduced_datatest) #### test data to predict
```

```python
Y_test.shape
```

```python
xgb_class = XGBRegressor(objective="reg:squarederror")
xgb_class.fit(X_train , Y_train)

y_pred = xgb_class.predict(Y_test) 

mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```
